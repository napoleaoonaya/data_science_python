```python
import pandas as pd
import plotly.express as px
import plotly.figure_factory as fig
import plotly.figure_factory as ff
import numpy as np
file = "C:\\Users\\Priscila\\Downloads\\netflix_titles.csv"
df = pd.read_csv(file)
df.head()
```

```python
df.dtypes
```

```python
filmes = df[df['type'] == 'Movie']
filmes.count()

```

```python
filmes['duration'].head()
```

```python
valor = int("90 min".split(" ")[0])
```

```python
def duration_numerica(valor):
    return valor
```

```python
duration_numerica(valor)
```

```python
filmes["duration_min"]=filmes['duration'].apply(duration_numerica)
```

```python
print(filmes["duration_min"])
```

```python
series = df[df['type']=='TV Show']
```

```python
series.count()
```

```python
series['listed_in'].value_counts()
```

```python
fig = px.bar(series['listed_in'].value_counts())
fig.update_layout(title_text="Bar Número de séries por genêro",
                 yaxis=dict(title_text="Quantidade"), xaxis=dict(title_text=""))
fig.show()
```

```python
generos_15 = series['listed_in'].value_counts()[series['listed_in'].value_counts() >= 15]
```

```python
fig = px.bar(generos_15, text = generos_15.values)
fig.update_layout(title_text="Bar Número de séries por genêro",
                 yaxis=dict(title_text="Quantidade"), xaxis=dict(title_text=""))
fig.update_traces(texttemplate='%{text:.2s}',textposition='outside')
fig.show()
```

```python
filmes["duration_min"] = pd.to_numeric(filmes["duration_min"], errors="coerce")
```

```python
filmes = filmes.dropna(subset=["duration_min"])
```

```python
if filmes["duration_min"].empty:
    print("No valid values in the 'duration_min' column.")
else:
    # Create the distribution plot
    f = ff.create_distplot([filmes["duration_min"]],
                           ['Duration min'], 
                           show_rug=True, bin_size=2, show_hist=True)
    f.update_layout(title_text="Duração dos filmes")
    f.show()
```
